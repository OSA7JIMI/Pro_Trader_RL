```python
import pandas as pd
from preprocessing import wrapper_func
```

```python
sp500 = pd.read_excel('SP500.xlsx')
sp500 = sp500.astype(str)
```

```python
data = pd.DataFrame()
buy_norm = pd.DataFrame()
sell_norm = pd.DataFrame()
for i in range(len(sp500)):
    row = sp500.iloc[i]
    symbol = row['Symbol']
    start = row['Date added']
    df, buy, sell = wrapper_func(symbol,start = start)
    data = pd.concat([data, df])
    buy_norm = pd.concat([buy_norm, buy])
    sell_norm = pd.concat([sell_norm, sell])
```

```python
data.to_csv('data.csv')
buy_norm.to_csv('buy_norm.csv', index = False)
sell_norm.to_csv('sell_norm.csv', index = False)
```

```python
above = buy_norm[buy_norm['signal_return'] > 10]
below = buy_norm[buy_norm['signal_return'] <= 10]
frac = len(above)/len(below)
if frac >1:
    frac = 1/frac
    df = pd.concat([above.sample(frac = frac), below]).sort_index().reset_index(drop=True)
else:
    df = pd.concat([above, below.sample(frac = frac)]).sort_index().reset_index(drop=True)
```
