```python
import pandas as pd
from preprocessing import wrapper_func
```

```python
from datetime import datetime
date_format = "%Y-%m-%d"
```

```python
data = pd.DataFrame()
buy_norm = pd.DataFrame()
sell_norm = pd.DataFrame()
```

```python
sp500 = pd.read_excel('SP500.xlsx')
sp500 = sp500.astype(str)

for i in range(len(sp500)):
    row = sp500.iloc[i]
    symbol = row['Symbol']
    dt1 = datetime.strptime(row['Date added'], date_format)
    dt2 = datetime.strptime('2015-01-01', date_format)
    start = max(dt1, dt2)
    df, buy, sell = wrapper_func(symbol,start = start,end = '2025-12-31')
    data = pd.concat([data, df])
    buy_norm = pd.concat([buy_norm, buy])
    sell_norm = pd.concat([sell_norm, sell])
```
