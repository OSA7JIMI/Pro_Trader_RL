```python
from stable_baselines3 import PPO
from stable_baselines3.common.env_util import make_vec_env
from environment import SellKnowledge
```

```python
n_envs = 4
lr = 0.0001
n_steps = 2048
batch_size = 64
ent_coef = 0.01
vf_coef = 0.5
n_epochs = 4622
```

```python
vec_env = make_vec_env(SellKnowledge, n_envs = n_envs)
model = PPO("MlpPolicy", vec_env, learning_rate = lr, n_steps = n_steps, batch_size = batch_size, ent_coef = ent_coef, vf_coef = vf_coef, verbose=1)
```

```python
model = model.learn(total_timesteps=n_steps*n_epochs)
```

```python
model.save('sell_knowledge_agent')
```
